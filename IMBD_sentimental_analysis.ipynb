```python
# Define the remove_stopwords function
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    filtered_text = ' '.join([word for word in text.split() if word.lower() not in stop_words])
    return filtered_text


# Apply remove_stopwords function to the DataFrame column
df['review'] = df['review'].apply(remove_stopwords)

# Display the DataFrame
print(df)
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to apply lemmatization
def apply_lemmatization(text):
    lemmatized_tokens = [lemmatizer.lemmatize(token, wordnet.VERB) for token in nltk.word_tokenize(text)]
    return ' '.join(lemmatized_tokens)
apply_lemmatization(df["review"][2])

# Apply stemming and lemmatization functions to the DataFrame column
df['review'] = df['review'].apply(apply_lemmatization)

# Display the DataFrame
print(df)
```

```python
# Function to remove numerical values from text
def remove_numerical_values(text):
    return re.sub(r'\d+', '', text)
# Apply remove_numerical_values function to the DataFrame column
df['review'] = df['review'].apply(remove_numerical_values)
```

```python
# Function to remove extra spaces from text
def remove_extra_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()

# Apply remove_extra_spaces function to the DataFrame column
df['review'] = df['review'].apply(remove_extra_spaces)

# Display the DataFrame
df[0:5]
```

```python
import pandas as pd

# Function to remove short words (2 letters or fewer) from text
def remove_short_words(text):
    return ' '.join([word for word in text.split() if len(word) > 2])

df['review'] = df['review'].apply(remove_short_words)

# Display the DataFrame
df[0:5]


```

```python
 #Tokenization function
def tokenize_text(text):
    return word_tokenize(text)

# Apply tokenization to the DataFrame column
df['review'] = df['review'].apply(tokenize_text)
```

```python
df1=pd.read_csv("clean_df.csv")
```

```python
df1=df1.drop(columns="Unnamed: 0")
```

```python
df1
```

```python
df1['sentiment'] = df1['sentiment'].replace({'positive': 1, 'negative': 0})
df1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
# Convert tokenized data to text
df1['review'] = df1['review'].apply(lambda tokens: ' '.join(eval(tokens)))
```

```python
df1
```

```python
# Initialize CountVectorizer
vectorizer = CountVectorizer(max_features=10000)
# Fit and transform the text data
X = vectorizer.fit_transform(df1['review'])
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, df1['sentiment'], test_size=0.2, random_state=42)

#print(X_train.toarray())
print(X_train.shape)

# Initialize Multinomial Naive Bayes classifier
classifier = MultinomialNB()

# Train the classifier
classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
